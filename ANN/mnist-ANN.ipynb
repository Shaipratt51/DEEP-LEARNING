```python
# 1. Import libraries
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.datasets import mnist
```

```python
# 2. Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 3. Normalize pixel values (0вЂ“255 в†’ 0вЂ“1)
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
# 4. Build ANN model
model = Sequential([
    Flatten(input_shape=(28, 28)),       # Flatten 28x28 image в†’ 784 vector
    Dense(128, activation='relu'),       # Hidden layer
    Dropout(0.3),                        # Regularization
    Dense(64, activation='relu'),        # Another hidden layer
    Dropout(0.3),
    Dense(10, activation='softmax')      # Output layer (10 classes: digits 0вЂ“9)
])
```

```python
# 5. Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# 6. Train model
history = model.fit(x_train, y_train, epochs=10,
                    validation_data=(x_test, y_test), batch_size=128)
```

```python
# 7. Evaluate model
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print("Test accuracy:", test_acc)
```

```python
# 8. Save model
model.save("mnist_ann_model.h5")
```
